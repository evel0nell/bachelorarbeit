```python
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("/Users/Thinkport/Documents/ba/bachelorarbeit/data/Experiments.csv", sep=';')
df.reset_index(inplace=True, drop=True)

del df["UID"]
del df["average "]
del df["step"]
del df["total_energy"]
del df["git repository"]
del df["Host OS"]
del df["Guest OS"]
```

```python
columns = [i.replace(" ", "_").lower() for i in df.columns]
```

```python
d = dict(zip(df.columns, columns))
```

```python
def transform_time(timestr):
    ftr = [3600,60,1]
    return sum([a*b for a,b in zip(ftr, map(int,timestr.split(':')))])
```

```python
df.rename(columns=d, inplace=True)
df
```

```python

df = df[df.energy.notna()]
df.energy = df.energy.apply(lambda x:float(x.replace(",", ".")))
df = df[~df.isnull().any(axis=1)]



df.time = df.time.apply(transform_time)
```

```python
df
```

```python
values_string = []
values_numbers = []

for col in df.columns:
    print("columns:", col)
    all_values = list(df[col])
    print(all_values)
    if type(all_values[0]) == str:
        values_string.append(col)
    else:
        values_numbers.append(col)
    print(set(df[col]))
    print()
```

```python
values_string = ['virtualization_technology']
```

```python
values_numbers
```

```python
from sklearn.preprocessing import normalize, OneHotEncoder
import numpy as np

df = df[df['status'] == 'success']
ohe = OneHotEncoder()
data = ohe.fit_transform(df[values_string])
df1 = pd.DataFrame(data.toarray(), columns=ohe.get_feature_names_out(), dtype=int)
df2 = pd.DataFrame(df[values_numbers].T, values_numbers).T
df_all = pd.concat([df1, df2, df.workflow], axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example dataset (replace this with your actual `df_all`)
# df_all = pd.read_csv('your_dataset.csv')

# Define a function to calculate average time for a given workflow
def calculate_average_time(group):
    return pd.DataFrame({
        'Technology': ['KVM', 'LXC', 'VirtualBox'],
        'Average Time (min)': [
            group.loc[group['virtualization_technology_KVM'] == 1, 'time'].mean() / 60,  # Convert seconds to minutes
            group.loc[group['virtualization_technology_LXC'] == 1, 'time'].mean() / 60,  # Convert seconds to minutes
            group.loc[group['virtualization_technology_VirtualBox'] == 1, 'time'].mean() / 60  # Convert seconds to minutes
        ]
    })

# Group by workflow and calculate averages
workflow_averages = df_all.groupby('workflow').apply(calculate_average_time).reset_index(level=1, drop=True)

# Visualization: Facet Grid for workflows
g = sns.FacetGrid(workflow_averages.reset_index(), col='workflow', sharey=False, height=5, aspect=1.5)
g.map_dataframe(sns.barplot, x='Average Time (min)', y='Technology', palette="flare", order=['KVM', 'LXC', 'VirtualBox'])

# Add labels to each bar
for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt="%.2f", label_type="edge", fontsize=10)

# Customize plot titles and axis labels
# Iterate over axes and set custom titles
custom_titles = ["Vue CI Workflow", "FreeCodeCamp Containers Workflow"]
for ax, title in zip(g.axes.flat, custom_titles):
    ax.set_title(title)
g.set_axis_labels("Average Time (min)", "Virtualization Technology")
g.fig.subplots_adjust(top=0.85)

# Save the plot
plt.tight_layout()
plt.savefig('/Users/Thinkport/Documents/ba/bachelorarbeit/results/timing_avrg3.png', dpi=600, bbox_inches='tight')
plt.show()
```
